## Exercise 1

**Follow along with these steps**
* Create a figure object called fig using plt.figure()
* Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis ax.
* Plot (x,y) on that axes and set the labels and titles to match the plot below:

```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
x = np.array([1,100])
ax.plot(x,2*x)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title')

plt.show()
```

## Exercise 4

**Use plt.subplots(nrows=1, ncols=2) to create the plot below.**

```python
import matplotlib.pyplot as plt
import numpy as np

plt.subplots(nrows=1, ncols=2)
```

**Now plot (x,y) and (x,z) on the axes. Play around with the linewidth and style**

```python
import matplotlib.pyplot as plt
import numpy as np

plt.subplots(nrows=1, ncols=2)
x = np.linspace(0,100)
plt.subplot(1,2,1).plot(x,2*x,linestyle="--", color="blue")
plt.subplot(1,2,2).plot(x,x**2, color = "red")
plt.show()
```

**See if you can resize the plot by adding the figsize() argument in plt.subplots() are copying and pasting your previous code.**

```python
import matplotlib.pyplot as plt
import numpy as np

plt.subplots(nrows=1, ncols=2, figsize=(15,3))
x = np.linspace(0,100)
plt.subplot(1,2,1).plot(x,2*x, color="blue")
plt.subplot(1,2,2).plot(x,x**2, color = "red",linestyle="--")
plt.show()
```
