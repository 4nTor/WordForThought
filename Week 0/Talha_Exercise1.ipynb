## Exercise 1

**Follow along with these steps**
* Create a figure object called fig using plt.figure()
* Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis ax.
* Plot (x,y) on that axes and set the labels and titles to match the plot below:

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
x=np.array([0,100])
y=np.array([0,200])
fig=plt.figure(figsize=(5.5,3.5))
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title')
```

**Now use x,y, and z arrays to recreate the plot below. Notice the xlimits and y limits on the inserted plot:**

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
fig=plt.figure(figsize=(5,3))
ax1=fig.add_axes([0,0,1,1])
ax2=fig.add_axes([0.2,0.5,.4,.4])
x=np.array([20.0,22.0])
y=np.array([40,44])
x2=np.linspace(0,10,100)
z=x2**2
ax1.plot(x2,z,label=r'$z = x^2$',color='b')
ax1.set_xlabel('X')
ax1.set_ylabel('Z')
plt.xlim(20.0,22.0)
plt.ylim(30,50)
ax2.plot(x,y,color='b')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('zoom')

```

**Now plot (x,y) and (x,z) on the axes. Play around with the linewidth and style**

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
x=np.array([0,100])
y=np.array([0,200])
x2=np.linspace(0,10,100)
z=x2**2
plt.figure(figsize=(5.5,3))
plt.subplot(1,2,1)
plt.plot(x,y, color='b',linestyle='dashed')
plt.subplot(1,2,2)
plt.plot(x2,z, label=r'$z = x^2$', color='r')


```

**See if you can resize the plot by adding the figsize() argument in plt.subplots() are copying and pasting your previous code.**

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
x=np.array([0,100])
y=np.array([0,200])
x2=np.linspace(0,10,100)
z=x2**2
plt.figure(figsize=(10,2))
plt.subplot(1,2,1)
plt.plot(x,y, color='b',linewidth='4.5')
plt.subplot(1,2,2)
plt.plot(x2,z, label=r'$z = x^2$', color='r', linestyle='dashed',linewidth='4.5')
```
