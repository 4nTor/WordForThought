```python
# Exercise 15
# Write a function definition named has_vowels that takes in value and returns True if the string contains any vowels.

def has_vowels(value):
    vowels = "aeiouAEIOU"
    for char in value:
        if char in vowels:
            return True
    return False

assert has_vowels("banana") == True, "Double check your syntax and logic"
assert has_vowels("ubuntu") == True, "Double check your syntax and logic"
assert has_vowels("QQQQ") == False, "Double check your syntax and logic"
assert has_vowels("wyrd") == False, "Double check your syntax and logic"

print("Exercise 15 is correct.")
```

```python
# Exercise 16
# Write a function definition named count_vowels that takes in value and returns the count of the number of vowels in a sequence.

def count_vowels(value):
    vowels = "aeiou"
    count = 0
    for char in value.lower():
        if char in vowels:
            count += 1
    return count

# Testing the function with various assertions
assert count_vowels("banana") == 3, "Double check your syntax and logic"
assert count_vowels("ubuntu") == 3, "Double check your syntax and logic"
assert count_vowels("mango") == 2, "Double check your syntax and logic"
assert count_vowels("QQQQ") == 0, "Double check your syntax and logic"
assert count_vowels("wyrd") == 0, "Double check your syntax and logic"

print("Exercise 16 is correct.")
```

```python
# Exercise 17
# Write a function definition named remove_vowels that takes in string and returns the string without any vowels

def remove_vowels(value):
    vowels = "aeiouAEIOU"
    result = ""
    for char in value:
        if char not in vowels:
            result += char
    return result

# Testing the function with various assertions
assert remove_vowels("banana") == "bnn", "Double check your syntax and logic"
assert remove_vowels("ubuntu") == "bnt", "Double check your syntax and logic"
assert remove_vowels("mango") == "mng", "Double check your syntax and logic"
assert remove_vowels("QQQQ") == "QQQQ", "Double check your syntax and logic"

print("Exercise 17 is correct.")
```

```python
# Exercise 18
# Write a function definition named starts_and_ends_with_vowel that takes in string and returns True if the string starts and ends with a vowel

def starts_and_ends_with_vowel(value):
    vowels = "aeiouAEIOU"
    return value[0] in vowels and value[-1] in vowels

assert starts_and_ends_with_vowel("ubuntu") == True, "Double check your syntax and logic"
assert starts_and_ends_with_vowel("banana") == False, "Double check your syntax and logic"
assert starts_and_ends_with_vowel("mango") == False, "Double check your syntax and logic"

print("Exercise 18 is correct.")
```
