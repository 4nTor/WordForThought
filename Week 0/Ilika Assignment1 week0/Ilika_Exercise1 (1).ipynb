## Data

```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

## Exercise 1

**Follow along with these steps**
* Create a figure object called fig using plt.figure()
* Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis ax.
* Plot (x,y) on that axes and set the labels and titles to match the plot below:

```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y, 'b')
ax.set_title('title')
ax.set_xlabel('x')
ax.set_ylabel('y')
plt.show()
```

## Exercise 2
**Create a figure object and put two axes on it, ax1 and ax2. Located at [0,0,1,1] and [0.2,0.5,.2,.2] respectively.**

```python
import matplotlib.pyplot as plt

# Create the figure and axes
fig = plt.figure()

# Add the first axis (ax1) at [0, 0, 1, 1]
ax1 = fig.add_axes([0, 0, 1, 1])  # Larger axis

# Add the second axis (ax2) at [0.2, 0.5, 0.2, 0.2]
ax2 = fig.add_axes([0.2, 0.5, 0.2, 0.2])  # Smaller axis
```

**Now plot (x,y) on both axes. And call your figure object to show it.**

```python
# Plot (x, y) on the first axis
ax1.plot(x, y, 'b')
ax1.set_title('Larger Plot')
ax1.set_xlabel('x')
ax1.set_ylabel('y')

# Plot (x, y) on the second axis
ax2.plot(x, y, 'r')
ax2.set_title('Smaller Plot')
ax2.set_xlabel('x')
ax2.set_ylabel('y')

# Show the figure
plt.show()


```

## Exercise 3

**Create the plot below by adding two axes to a figure object at [0,0,1,1] and [0.2,0.5,.4,.4]**

```python
import matplotlib.pyplot as plt

# Create a figure
fig = plt.figure()

# Add first axis at position [0, 0, 1, 1] (taking up the full figure)
ax1 = fig.add_axes([0, 0, 1, 1])  # [left, bottom, width, height]
ax1.set_title("Axis 1 (Full figure)")

# Add second axis at position [0.2, 0.5, 0.4, 0.4] (smaller, offset)
ax2 = fig.add_axes([0.2, 0.5, 0.4, 0.4])  # [left, bottom, width, height]
ax2.set_title("Axis 2 (Smaller region)")

# Display the figure
plt.show()

```

**Now use x,y, and z arrays to recreate the plot below. Notice the xlimits and y limits on the inserted plot:**

```python
fig = plt.figure()

ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.2, 0.5, 0.4, 0.4])

# Larger plot
ax1.plot(x, z, 'b')
ax1.set_title('Larger Plot')
ax1.set_xlabel('x')
ax1.set_ylabel('z')

# Smaller plot with limits
ax2.plot(x, y, 'r')
ax2.set_title('Smaller Plot')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_xlim(20, 22)
ax2.set_ylim(30, 50)

plt.show()
```

## Exercise 4

**Use plt.subplots(nrows=1, ncols=2) to create the plot below.**

```python
import matplotlib.pyplot as plt

# Create a figure with 1 row and 2 columns of subplots
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2)

# Set the limits for both axes from 0 to 1
ax1.set_xlim(0, 1)
ax1.set_ylim(0, 1)
ax1.set_title("Axis 1 (Left)")

ax2.set_xlim(0, 1)
ax2.set_ylim(0, 1)
ax2.set_title("Axis 2 (Right)")

# Display the figure
plt.show()
```

**Now plot (x,y) and (x,z) on the axes. Play around with the linewidth and style**

```python
import matplotlib.pyplot as plt

# Create subplots with one row and two columns, and set the figure size
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))

# First subplot: Plot (x, y)
axes[0].plot(x, y, linestyle='--', color='b', linewidth=2)
axes[0].set_title('Plot 1')
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')

# Second subplot: Plot (x, z)
axes[1].plot(x, z, linestyle='-', color='r', linewidth=3)
axes[1].set_title('Plot 2')
axes[1].set_xlabel('x')
axes[1].set_ylabel('z')

# Adjust layout to prevent overlapping
plt.tight_layout()

# Show the plots
plt.show()


```
