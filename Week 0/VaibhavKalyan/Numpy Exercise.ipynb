# NumPy Exercises

#### Import NumPy as np

```python
import numpy as np
```

#### Create an array of 10 zeros

```python
import numpy as np

array = np.zeros(10)

print(array)
```

#### Create an array of 10 ones

```python
import numpy as np

array = np.ones(10)

print(array)
```

#### Create an array of 10 fives

```python
import numpy as np

array = np.ones(10) * 5

print(array)
```

#### Create an array of the integers from 10 to 50

```python
array = list(range(10, 51))
print(array)
```

#### Create an array of all the even integers from 10 to 50

```python
array = list(range(10, 51, 2))
print(array)

```

#### Create a 3x3 matrix with values ranging from 0 to 8

```python
import numpy as np

matrix = np.arange(9).reshape(3, 3)
print("array")
print(matrix)
```

#### Create a 3x3 identity matrix

```python
import numpy as np

matrix = np.eye(3)
print("array")
print(matrix)
```

#### Use NumPy to generate a random number between 0 and 1

```python
import numpy as np

num = np.random.rand()
print(num)
```

#### Use NumPy to generate an array of 25 random numbers sampled from a standard normal distribution

```python
import numpy as np

random = np.random.randn(25)
print("array")
print(random)
```

## Numpy Indexing and Selection

Now you will be given a few matrices, and be asked to replicate the resulting matrix outputs:

```python
import numpy as np
matrix = np.arange(1, 26).reshape(5, 5)
print("array")
print(matrix)
```

```python
import numpy as np
matrixfrom12to25 = np.arange(1, 26).reshape(5, 5)
print(matrixfrom12to25[2:, 1:])
```

```python
import numpy as np

matrix = np.arange(1, 26).reshape(5, 5)


print( matrix[3, 4])
```

```python
import numpy as np

array = np.arange(1, 26).reshape(5, 5)
array[[0, 1, 2], 1:2]
print("array")
print(array[[0, 1, 2], 1:2])
```

```python
import numpy as np
matrix = np.arange(1, 26).reshape(5, 5)
print("array")
print(matrix[4:])
```
